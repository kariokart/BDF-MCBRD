```python
###Python code for blurring pictures###

#Load necessary modules 
import skimage
from skimage.viewer import ImageViewer
import psutil

#open list of images

with open('dirlist.txt') as list:
    for line in list:
        name = line
        name = name.rstrip('\n')

    # read and display original image
        image = skimage.io.imread(fname=name)
       # viewer = ImageViewer(image)
      #  viewer.show()
        

    #Set sigma and blur the picture
        sigma = float(30.0) 

        blurred = skimage.filters.gaussian(
            image, sigma=(sigma, sigma), multichannel=True)

    # display and save the new blurred image
        #viewer = ImageViewer(blurred)
        #viewer.show()
        saveName = 'blurred' + name
        #blurred.imsave(saveName)
        #viewer.save_to_file(saveName)
        skimage.io.imsave(fname=saveName, arr=blurred)


```
